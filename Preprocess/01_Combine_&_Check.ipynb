# CoAID

Combine Files with Labels

```python
ctweet_nolabel = pd.read_csv("CoAID_ids_complete.csv")
ctweet_nolabel.shape #5808*35
ctweet_label = pd.read_csv("CoAID_ids_labelled.txt",delimiter=',')
ctweet_label #6799*3
```

```python
ctweet = pd.merge(ctweet_nolabel, ctweet_label, how='outer', on='id')
ctweet.shape #7865, 37
ctweet[['text','is_misinfo']].head()
```

**CoAID dataset didn't label the sarcastic content correctly. Just like the first example, 'Excellent, devastating story. Anyone who says covid-19 is “just like the flu” needs to read this https://t.co/c1UaBU9lU1'. This was detected as misleading information which we know it is not true. I decided not to use this dataset.**
